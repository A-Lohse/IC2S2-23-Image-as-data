### THINKING BREAK 2: What are images acutally?

---
We often don't pause to consider: What are images actually composed of? Well, in the digital world, images are typically represented as multidimensional arrays, i.e. tensors, of pixel values. This means that images can be visualized as stacked matrices of pixel intensity values, which determine the appearance of the final image. You can imagine it as an Excel spreadsheet with different "panes" stacked on top of one another, as shown in the figure below. An image consists of what we can call height, width, and channel dimensions. The height and width are self-explanatory, but the channel dimension requires some explanation. The channels of an image represent the number of color channels the image has. For instance, a grayscale image consists of 1 channel (white/black), whereas most color images in the RGB format consist of red (R), green (G), and blue (B) channels. Some images may also have a 4th channel determining the transparency of the colors. Each entry in these channels corresponds to a pixel value or intensity (Note that in theory, you use the models below on tensors with arbitrary numbers of channels).

This may not be the way humans typically perceive images, but it is how computers process and analyze them. Understanding this concept is essential when working with computational tools.

In this exercise, we will explore the creation of an *Image Classifier* using both a *dense* Artificial Neural Network (ANN) and a Convolutional Neural Network (CNN). While the dense ANN is a "standard" neural network, the CNN is specifically designed to learn complex spatial patterns found in images.

CNNs are particularly powerful models for Image Classification tasks. When utilized as an *Image Classifier*, the CNN aims to label images based on their content. This can be done in a single-label setting or a multi-label setting, as well a multi class og single-class setting. Let's take a moment to grasp the meaning of that sentence.

Multi and single-class refers to the number of "classes" or labels, that a models can choose from. A single class task could determening wheter an image is of [of a hot dog or not a hot dog](https://www.youtube.com/watch?v=tWwCK95X6go) or maybe a selfie or not. Multiclass is the more common taks, and refers cases where you have more classes such as gender, facial recocnition, animal identification etc.

Indepentent of how many classes we we also have the destinction between a single-label and multi-label setting. In a single-label setting, our model will produce only one label per image. This is useful in cases where images can belong to only one of several exclusive classes (for example, an image can be either a selfie or not, or an image of one specific bird species among many others).

In contrast, multi-label classification expects the model to output *all* the objects present in the image. Note that multi-label tasks can always be reduced to a series of binary single-label tasks. We can combine single/multi label and class as need.


For simplicity, we will focus on the single-label multiclass setting in this exercise. Meaning we want our model to make a single prediction from a list of classes.

**Transformations**

As we load the data, we will also apply a couple of transformations to the images. These are, as the name implies, simply changes that we apply to all the images, some of which are know to help us train the models and some are simply to tranform them to the rigth format.  

To achieve this, we utilize the `transforms.Compose` function provided by PyTorch. Within this composition, we can include all the transformations we desire. In this case, we will apply two specific transformations: `transforms.ToTensor()` and `transforms.Normalize()`. These transformations serve two purposes: converting the numpy arrays into PyTorch tensors and normalizing the RGB values, which range from 0 to 255, to a new range of 0 to 1.


**transforms.ToTensor:** This transformation converts the images into tensors, which are the fundamental data structure used in PyTorch. Tensors are multidimensional arrays that contain elements of a single data type, similar to numpy arrays. However, the key distinctions are that PyTorch tensors can utilize GPUs for faster numerical computations and offer automatic differentiation for constructing and training deep learning models, enabling gradient-based optimization algorithms. Additionally, this transformation scales the image pixel values to a range of [0, 1]. You can read more about pytorch tensors [here](https://pytorch.org/docs/stable/tensors.html#)

**transforms.Normalize()**: This transformation is used to normalize the pixel values of an image. To apply this transformation, we need to specify the mean and standard deviation for each channel of the image. Since we are working with color images, which consist of three channels (red, green, and blue), we need to perform this normalization process separately for each channel. Considering that the `transforms.ToTensor()` function scales the image values to the range of [0, 1], we can set both the standard deviation and mean to 0.5 for all three channels. This normalization ensures that the images have a range of approximately [-1, 1], centered around 0. By performing this normalization, we contribute to the convergence and stability of our neural network during the training phase.

Note that the same transformations are applied to both the training and test sets. This includes converting the images to tensors and normalizing them. However, it is possible to apply additional transformations, such as rotations, cropping, scaling, and more, specifically to the training set. These additional transformations can help mitigate the issue of overfitting. For more information on transformations, you can refer to the documentation [here](https://pytorch.org/vision/stable/transforms.html).


**DataLoaders**

Loading images or other unstructured data into your computer's memory can be resource-intensive. To mitigate this, we typically use a *DataLoader*, which allows us to load data incrementally, piece by piece. With a DataLoader object in PyTorch, we can load references to our dataset and load the actual data into memory only when we need a specific observation. This approach prevents overburdening the machine you are working on. Additionally, a DataLoader can be used to randomly sample from the data, apply transformations, and more. Overall, it is a very useful tool. However, when you want to perform operations on the entire dataset, you will need to loop over the DataLoader and apply the desired actions.

This is a rather clever solution, but using a DataLoader may feel unfamiliar to those accustomed to working with tabular datasets. While it makes sense to create a custom DataLoader class for many tasks, in this exercise, we will use the default DataLoader provided by PyTorch.

### EXPERT NOTES ON BATCH SIZES


---
**Large Batch Sizes:**


*Pros:*

- Faster Training: With larger batch sizes, parallelization across multiple processors or GPUs can be more efficient, leading to faster training times.
- Smoothed Gradients: Computing gradients based on a larger number of samples can provide a smoother estimation of the true gradients, which can lead to more stable updates and better convergence.
- Utilization of Hardware: Training with larger batch sizes allows for better utilization of hardware resources, especially when using specialized hardware like GPUs.

*Cons:*
- Memory Requirements: Larger batch sizes require more memory to store the intermediate activations and gradients, which can be a limitation on devices with limited memory capacity.
- Generalization: In some cases, using very large batch sizes can result in models that generalize poorly, as the model may not encounter enough diversity in the data within each batch to learn effectively. This can lead to overfitting on the training data and poorer performance on unseen data.
- Limited Exploration: Large batch sizes may reduce the diversity of samples within each batch, potentially limiting the exploration of different patterns in the data.

**Small Batch Sizes:**

*Pros:*
- Better Generalization: Smaller batch sizes can help models generalize better as they encounter a wider range of samples and have more opportunities to update the model parameters based on different patterns in the data.
- Memory Efficiency: Using smaller batch sizes requires less memory, making it more feasible to train models with limited resources or on devices with lower memory capacity.
- Exploration: Smaller batch sizes enable more diverse samples within each batch, allowing the model to explore different regions of the data distribution.

*Cons:*
- Slower Training: Training with smaller batch sizes typically requires more iterations to process the entire dataset, which can lead to slower training times, especially if parallelization across multiple processors or GPUs is not fully utilized.
- Noisy Gradients: Computing gradients based on a small number of samples can result in noisy estimations of the true gradients, which may introduce more randomness and make convergence less stable.
- Hardware Utilization: Using small batch sizes may result in underutilization of hardware resources, particularly if the hardware is capable of processing larger batches efficiently.

Let's quickly explain what a neural network is and how it can be trained to classify images and perform other tasks. While we would love to delve into more detail, it is beyond the scope of this tutorial. However, we can provide a brief overview.

Firstly, a neural network can be viewed as a universal function approximator that maps an input x to an output y using parameters, or weights, $$:

$y = f(x; )$

What sets neural networks apart from most other machine learning approaches is their depth. This means that the parameters  are distributed across multiple hidden layers. We can understand a network as composed of many embedded functions:

$y = f^(3)(f^(2)(f^(1)(x; ^(1)); ^(2)); ^(3))$

Thus, the original input x is transformed repeatedly through each layer. To enable the creation of highly complex functions, each layer is followed by a non-linear activation function.

The final output layer transforms the input into a prediction of the output (). For image classification, this could be the probability of an image containing a cat, dog, boat, etc.

One central aspect is that the hidden layers of the neural network enable the model to automatically learn and generate features that are relevant for mapping inputs (x) to outputs (y). As we delve deeper into the network, it becomes more proficient at discovering and creating increasingly complex features. More on this later.

###EXPERT NOTES ON THE OUTPUT LAYER
---


The keen observer will note that for a classification task (with +2 classes), the output layer should be followed by a softmax activation function to nomalize the "logits" to probabilities $\in [0,1]$. Yet no such final activation is present above.

The reason for avoiding the direct conversion of raw output logits to probabilities and subsequently using them with the cross-entropy loss is due to practical computational considerations. While such conversion works perfectly well in theoretical mathematics, it introduces unnecessary complexity and can lead to numerical instability in practical computations. One issue arises from computational underflow, where very low probabilities are rounded down to zero, resulting in undefined logarithmic calculations. Since the cross-entropy loss requires taking the logarithm of predicted probabilities, the logarithm of zero leads to an undefined result. In Python, attempting to calculate log(0) will yield a value of negative infinity (-inf).

The [nn.CrossEntropyLoss()](https://pytorch.org/docs/stable/generated/torch.nn.CrossEntropyLoss.html) function we will use in the code combines the softmax activation and the negative log-likelihood loss into a single step. By bypassing the need for a separate softmax activation in the network's output, we can directly utilize the logits as inputs to the CrossEntropyLoss. This approach offers improved computational efficiency and avoids potential numerical problems that can arise from exponentiation in softmax calculations.

Note, however, that if you ever need to implement your own loss functions, such as the  [focal loss](https://openaccess.thecvf.com/content_iccv_2017/html/Lin_Focal_Loss_for_ICCV_2017_paper.html) for handling class imbalance, you may need to include a softmax activation function after your final layer to ensure valid probabilities, depending on your implementation. In such cases, a common practice is to replace "zero" predictions with a very small positive value, such as e^(-100), to prevent numerical instability when taking the logarithm of zero.

Hopefully, you found the above approach, "DenseNet," to work. Converting an image to a 1D vector and training a network on such vectors does yield results.

However, using a fully connected network to process this vector representation is not very efficient. Firstly, the input layer requires a separate parameter for each pixel in the input image. This results in a large number of weights to train, especially for larger images. Additionally, treating each pixel as an individual entity is not the most effective use of information given the nature of image data.

For instance, if we return to the example of the hand-drawn digit: imagine shifting the position of the 7 slightly. Just a couple of pixles in one or tow directions. You would likely not notice a difference, but this small change would result in a very different "barcode" representation for the model to learn.

Now, consider a more complex task: classifying an image as either "hot dog" or "not a hot dog." Any appearance of a hot dog in the image would help us make this distinction. However, when using a dense network, the learned patterns would be location-specific. In other words, if we encounter an image with a hot dog in a completely new position, our model may not recognize it as a hot dog, even if it has seen similar patterns thousands of times in other positions and locations.


As such, we would like a network architecture that is more efficient: a type of network that enables patterns to be independent of position and location. In other words, the content of the image should be able to "move" without adversely affecting the performance of our model. This leads us to CNNs.

The central component of a CNN is the convolutional layer, which performs a convolutional operation on a 2D grid with multiple channels. This operation is defined by several key elements:

- A square kernel/filter: A small square filter that is applied to the input grid.
- Multiple kernels/filters: There can be multiple kernels/filters applied to the input grid, each capturing different features.
- Stride: Determines the step size at which the kernel/filter is moved across the input grid.
- Optional padding: Padding can be added to the input grid to preserve spatial dimensions.

For a more detailed understanding of convolutional layers, we recommend revisiting the "fundamental" readings mentioned earlier. In summary, a convolutional layer applies a kernel, which is a small square filter containing the parameters/weights, to the input grid. By sliding the filter over the input grid from left to right and top to bottom, we calculate the weighted sum at each position. This process is illustrated [here](https://learnopencv.com/wp-content/uploads/2017/11/convolution-example-matrix.gif)


It should be clear from the GIF that the size of the output, also known as the feature map, is determined by the input size and the filter size. When considering only these hyperparameters, a simple equation can be used to calculate the size of the output or feature map dimension.

Width $ = W$

Height $ = H$  

Filter size = $F_{w}, F_{h}$  

Output width $ = W-F_{w} + 1$

Output height $ = H-F_{H} + 1$

Since we usually (but not always) work with squares, we often have W=H.

We always work with square kernels so $F_{W} = F_{H}$

What does our kernel capture? And what is the dimension of the resulting feature map?

By using the given kernel, you have detected vertical edges. As the filter traversed the "image," it identified all the vertical edges present. Therefore, the resulting feature map indicates the locations in the original "image" where vertical edges are found.

In CNNs, the individual weights of the kernel are not manually designed but learned through forward and backpropagation. Typically, multiple kernels or filters are applied to each image, resulting in several new feature maps.

Considering the aforementioned hard-coded example, we could envision having filters for vertical edges, horizontal edges, diagonals, and so forth.

So what happens when we go deep? We refer to it as a Convolutional Network when we employ the convolution operation instead of general matrix multiplication (as done in dense layers) in at least one of the layers. Technically, there is no difference between running a convolutional kernel over the original input image and running it over the generated feature map. The resulting feature maps are simply alternative (latent) representations of the images. Therefore, we can have multiple layers of convolutions, each operating on the feature map from the preceding layer (or the image itself in the case of the first layer).

When we use a deep CNN encoder architecture with non-linear activation functions in each layer, we enable the model to uncover increasingly complex and abstract patterns or features as we progress through the layers. Let's break it down step by step.

In the beginning, the initial layers of the CNN focus on detecting simple, yet general, features like edges and corners. They examine small parts of the image or feature map using small receptive fields, which represent the area covered by the kernel. As we move deeper into the network, the feature maps become smaller, and the kernels become relatively larger. This enlargement allows the kernels to cover a larger receptive field, capturing more contextual information.

As we reach the middle layers, more complex, yet still general, structures such as polygons and textures are captured. These layers develop a better understanding of the relationships between different elements in the image, allowing them to perceive larger patterns and identify more intricate features.

In the deepest layers of the network, the layers specialize in recognizing intricate objects like eyes, wheels, faces, crowds, and more. These layers learn to detect specific high-level features that are characteristic of the objects we want to identify (See Zeiler and Fergus 2013 for a visualization).

To retain the relevant information from the original image, we need to increase the number of channels. Channels refer to the different feature maps that are produced by the convolutional layers. By expanding the number of channels, the network can preserve important details and capture a richer representation of the image, enhancing its ability to recognize and distinguish various objects and patterns.

By combining these architectural principles and techniques, the deep CNN encoder can effectively extract and learn hierarchical representations of the input data. That is, the network progresses from capturing basic geometric features (color intensity and location) to extracting more semantic features. Thus enabling more powerful and accurate image analysis and classification.

We can decrease the height and width of an image or feature map in roughly two ways:

1. Valid Convolutions: One approach to shrinking the width and height of an image or feature map is through valid convolutions. This means that the size of the output feature map is determined solely by the input feature map and the stride (which we will discuss later). Valid convolutions are a natural approach as they are inherent to the convolutional operation. However, they can be inflexible when creating deep networks to capture increasingly complex patterns. Without control, the feature map could shrink too quickly, resulting in a tiny feature map of just one pixel.

2. Same Convolutions: To address the issue of rapid shrinking in valid convolutions, we introduce two additional components of the convolutional operation: stride and padding. Stride determines how many pixels the kernel moves at each step, with a default value of 1 for convolutions. Padding refers to artificially added pixels (often zeros) around the image or feature map. By adjusting the hyperparameters of kernel size, stride, and padding, we can ensure that the output feature map retains the size of the input feature map. The amount of padding required for "same" convolutions depends on the kernel size ($f_{W}, f_{H}$) and stride ($S_{W}, S_{H}$).

Therefore, a more comprehensive equation for the dimensions of the output feature map can be expressed as follows:

Output width$ = \frac{W - F_{w} + 2P}{S_{W}} + 1$

Output height$ = \frac{W - F_{H} + 2P}{S_{H}} + 1$

By alternating between valid and same convolutions, we can create arbitrarily deep networks where the feature map decreases in size at an appropriate rate. However, in practice, the shrinkage of feature maps is typically achieved using pooling rather than valid convolutions.

There are two types of pooling:

1. Max Pooling: Max pooling involves finding the maximum value within a sliding filter window (with a specified dimension) over a feature map and returning it in a new matrix.

2. Average Pooling: Average pooling computes the average value within the filter window over the feature map and returns it in a new matrix.

Both pooling methods serve the same purpose: reducing the size of the feature map. Max pooling retains the strongest signal, while average pooling retains the average signal. Pooling operations speed up computation, enhance robustness of detected features, and increase the receptive field. In practice, 2x2 max pooling with a stride of 2 is often used, resulting in an output feature map with half the height and half the width of the input feature map, simplifying the "bookkeeping" process.

**Advantages of CNNs**

All the atributes presented above, meand that Convolutional Neural Networks (CNNs) offer several advantages over dense networks when working with images:

- **Local Connectivity:** CNNs exploit the spatial structure of images by using local connectivity patterns. Unlike dense networks, which connect every neuron to every other neuron in the adjacent layers, CNNs use convolutional layers to only connect neurons within a receptive field. This local connectivity allows CNNs to capture local patterns and spatial relationships in images effectively.

- **Parameter Sharing:** CNNs utilize parameter sharing, where the same set of weights is used across different spatial locations in the image. This sharing of weights reduces the total number of parameters in the network, making it more memory-efficient and reducing the risk of overfitting. It also enables CNNs to learn hierarchical features at different levels of abstraction.

- **Translation Invariance:** CNNs inherently possess translation invariance, which means they can recognize patterns regardless of their location in the image. This property is achieved through the use of pooling layers that downsample the feature maps, capturing the most salient information while reducing sensitivity to small shifts or translations in the input image.

- **Hierarchical Feature Learning:** CNNs are designed to learn hierarchical representations of features. The initial layers in a CNN learn simple low-level features like edges and gradients, while deeper layers learn more complex and abstract features. This hierarchical learning allows CNNs to capture both local and global image features, leading to improved performance in image recognition tasks.

- **Efficiency:** Due to the local connectivity and parameter sharing, CNNs are computationally efficient compared to dense networks. The shared weights and sparse connectivity make CNNs well-suited for processing large image datasets efficiently.

Overall, the advantages of using CNNs make them highly effective for grid/image-related tasks, such as image classification, object detection, and image segmentation, by leveraging the inherent properties and structure of images.

In the context of convolutional neural networks (CNNs) and image processing, the inductive bias arises from the understanding that images have certain properties and structures. These properties include local patterns and spatial relationships, where neighboring pixels tend to be more related to each other than pixels that are farther apart. This bias allows CNNs to exploit the locality and spatial dependencies present in images.

However, inductive bias can also introduce limitations and biases that don't necessarily match the true patterns in the data. To overcome this challenge, transformer networks have been developed for image processing, specifically designed to "assume less". However, a trade-off exists: transformer networks typically require a larger dataset to learn what is implicitly enforced by using CNNs.

### THINKING BREAK 6 - EVALUATION AND METRICS

Accuracy is not the only metrics worth looking at. For tasks like image labelling it *can* be usefull, but there are also cases where it is not. Consider for example a case where you have an unbalanced dataset, with lets say 90% of one class. In this case you could obtain an accuracy og 0.9, by simply always predicting the largests class.

For this, and other reasons, we also use metrics like *recall*, *precision* and *F1* score which all consider a combination of false/true positive/negatives differently. For an overview we recomend you briefly check out the *classification metrics* section [here](https://neptune.ai/blog/performance-metrics-in-machine-learning-complete-guide)

In some cases it makes sense to optimize a specific metric. We may for example ensure that we miss as few true labels as possible, and don't care as much if we have a few more false positives. This may be the case in medical images analysis, where we would rather flag to many lung scans for review, as oppose to falsely proclaiming somene to be well. In this case we could chose the model with the best recall, as this is a metrics for specifically that.

Things get more complicated when we have multilabels settings, and when we move away from simple image labelling, so we need to be mindful about the metrics we optimize for and report, as they may mislead us.

### THINKING BREAK 7 - What is fine-tuning and why is it amazing

---

**What is finetuning**

Finetuning refers to the process of taking a ANN that has already been trained to a task, and re-training it (sligthly) to a new task. This is also sometimes known as transfer learning. The underlying conceptual idea is that for any training task, be it images, text ect. our ANN will learn both specific patterns of the training data, as well as more general patters of for example images, that all image models must learn anyway. With finetuning we want to utilize this, and take models that have already been trained on other data, and re-train them on our data to our task. In theory any model can be used as a base model, but the close the training taks and the data of the base model are to your task, the more transferable knowledge you can probably gain.

For this reason considerable effort has gone into making good and general base models that are trained on a lot of different data, and can thus be used to transfer learn many different tasks. Finetuning these models most often perform better than models we can train for our specific task.

Finetuning generalize base models is a very common technique, especially the field of NLP where large models like BERT, RoBERTa, and GPT3. are often used as "base models" to finetune. While seem very intuitive that there are general patterns in langugage that you can learn, it may not be as ituitive with images. However, it turns out that if you train ANNs on enough image data, the networks can also learn general patterns in images.

**How is finetuning done**

The technical way finetuning is done, is by changing the output/classification layer of the base model or adding a full new one to make it fit your task, and then keep training the model on your own data. If for example the base model has been trained to to classify images into 1000 classes, but you only need 10, well then you need to either add a new final layer or change the one that is already there. The actual training in the finetuning can generally be done in two ways:

1. Finetune the full base network (takes the longet but often gives better results)
2. Finetune only the final classification layer. This is generally refered to as freezing the layers of the model

Both approaches a completely valid, and often it is wise to expereiment with them. You can of couse also combine as you wish, to make more layers trainable than just the last. If you want more of how finetuning works, you can also whatch this [video explainer](https://www.youtube.com/watch?v=3gyeDlZqWko)

**Advantages of finetuning**
1. Extremely effective (fast and good)
2. Requires way less (labelled) data
3. Leaves most of the training decisions to ML/AI proffessionals
4. Only gets better as better base models are produced all the time


**Disadvantages of finetuning**
1. It does not always work
2. You do not (always) know the data the models have been trained on
3. Less flexible than making your own model

### THINKING BREAK 8 - What is Generative models, what can we use them for, and what is zero shot learning
---

In this section we examine one of the most recent advances in computer vision. That is generative models. These are deep learning models, that are able to generate images, text, etc based on the input they get. The most notable of such models rigth not is probably ChatGPT, which you probably know. It is beyond this tutorial to go into depth with how generative models work, but the interested reader can examine some of the papers in Inspirationa Research 4. Here we will give an idea of the intuition behind generative models, and how we can use them.

The most popular and modern generative architecture is called a diffusion model. These models are trained by adding more and more random noise to images, and then making the model recunstruct the orginal image, punishing it for wrongly reconstructing the image. This allows a trained model to produce immages from noice (or text prompts etc.).

Other popular models include variational autoencoders which work by making a dense encoding space or low dimension representation of the training data, and then recunstructing the images from this space. When generating new images, we then simply need to sample from the latent space.

finally there are also generative advasarial networks, which pit two networks against each other. One that generates new images, and one that tries to determince what is a true image and what is generated. We then train them my making it as dificult as possible to determine the true image. When wanting to generate new images, we just use the generator network to generate images.

All of these models can then be trained further, in order to map text or images to the generation process. In this way you can get models where you can pass a string of text as an input and get an image as an output. Or in other words, you can "ask" the model in natural language to "please make a picture of the pope in a fur coat", and it will produce something resonable.

Generative models do not have to be *pre-trained*, but most are now-a-days. You can train one from scratch yourself, but it takes a huge effort and a lot of data. Therefore most people rely on the very large pre-trained generative models already on the market today. One of the most prominent text-to-image models is the socalled "stabel-diffuser" model, which we will familiarize ourself with in the next coding exercise.

Generative models can of course be used to generate images, which can seem like a pretty gimmiky thing. But try to be creative, and see if you can't come up with some use cases for generating images in your research.

A second use case of generative models is what is called zero (or few) shot learning. Here we use an image-to-text model, to give us labels for what is in the image. We can even restrict the model, to only pick from a list of words. While this is conceptually a different task than image classification, it produces the same results (labels for each image), and can sometimes even give better results than homemade or finetuned models.
